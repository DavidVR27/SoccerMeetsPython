```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("PremierLeagueGK_Stats.csv")
```

```python
df.head()
```

```python
# o -> Puntos del plot
plt.plot(df['SoTA'],df['Save%'],'o')
```

```python
# Creando area del plot
fig, ax = plt.subplots()

# TamaÃ±o del plot
fig.set_size_inches(5,5)

plt.plot(df['SoTA'],df['Save%'],"*")

# Labels
ax.set_title("Shots Against vs Save %")
ax.set_xlabel("SoTA")
ax.set_ylabel("Save%")

plt.show()
```

```python

fig, ax = plt.subplots()
fig.set_size_inches(7,5)

# Scatter plot
plt.plot(df['SoTA'], df['Save%'], 'o')

# LÃ­neas de cuadrantes usando los lÃ­mites reales de los datos
plt.plot([df['SoTA'].mean(), df['SoTA'].mean()],
         [df['Save%'].min()-1, df['Save%'].max()+1], 'k-', linestyle="--", lw=1)

plt.plot([df['SoTA'].min()-5, df['SoTA'].max()+5],
         [df['Save%'].mean(), df['Save%'].mean()], 'k-', linestyle="--", lw=1)

ax.set_title("Shots on Target Against vs Save %")
ax.set_xlabel("SoTA")
ax.set_ylabel("Save%")

plt.show()
```

```python

fig, ax = plt.subplots()
fig.set_size_inches(7,5)

# Scatter plot
plt.plot(df['SoTA'], df['Save%'], 'o')

# LÃ­neas de cuadrantes usando los lÃ­mites reales de los datos
plt.plot([df['SoTA'].mean(), df['SoTA'].mean()],
         [df['Save%'].min()-1, df['Save%'].max()+1], 'k-', linestyle="--", lw=1)

plt.plot([df['SoTA'].min()-5, df['SoTA'].max()+5],
         [df['Save%'].mean(), df['Save%'].mean()], 'k-', linestyle="--", lw=1)

ax.set_title("Shots on Target Against vs Save %")
ax.set_xlabel("SoTA")
ax.set_ylabel("Save%")

ax.text(110,76,"Less shots on target, higher save %", color="black", size=8)
ax.text(190,62,"More shots on target, lower save %", color="black", size=8)

plt.show()
```
