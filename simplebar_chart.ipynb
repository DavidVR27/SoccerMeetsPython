```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Lectura y limpieza de datos
table = pd.read_csv("data/season-tables/premtable_2016-17.csv", header = 1)
table.index = np.arange(1, len(table) + 1)
table = table.drop(columns = ['Unnamed: 0'])
table = table.rename(columns = {'Unnamed: 1':'Team'})

# Reordenando columnas
total_league_stats = ['P.2', 'W.2', 'D.2', 'L.2']
first_column = [table.columns[0]]
other_columns = [col for col in table.columns[1:] if col not in total_league_stats]
table = table[first_column + total_league_stats + other_columns]
```

```python
# Renombrado de tablas
mapeo_columnas = {
    'P.2': 'P',
    'W.2': 'W',
    'D.2': 'D',
    'L.2': 'L',

    'P': 'P_home',
    'W': 'W_home',
    'D': 'D_home',
    'L': 'L_home',

    'P.1': 'P_away',
    'W.1': 'W_away',
    'D.1': 'D_away',
    'L.1': 'L_away'
}

table.rename(columns=mapeo_columnas, inplace=True)

table.head()
```

```python
# Grafico
plt.bar(x=np.arange(1,21),height=table['Pts'])

# Titulo del grafico
plt.title("Premier League 16/17")

# Dandole a las aristas un titulo
plt.xlabel("Teams")
plt.ylabel("Points")

# Nombres por equipo
plt.xticks(np.arange(1,21), table['Team'], rotation=90)

plt.show()
```

```python
# Arreglo de colores
teamColours = ['#034694','#001C58','#5CBFEB','#D00027',
              '#EF0107','#DA020E','#274488','#ED1A3B',
               '#000000','#091453','#60223B','#0053A0',
               '#E03A3E','#1B458F','#000000','#53162f',
               '#FBEE23','#EF6610','#C92520','#BA1F1A']

plt.bar(x=np.arange(1,21),height=table['Pts'], color=teamColours)
plt.title("Premier League 16/17")
plt.xticks(np.arange(1,21), table['Team'], rotation=90)

plt.xlabel("Teams")
plt.ylabel("Points")
plt.show()
```
